```python
import pandas as pd
```

```python
data=pd.read_html('https://www.espn.co.uk/olympics/summer/2020/medals')
```

```python
df=data[3]
df
```

```python
df1 = df.drop('Total', axis=1)
df1
```

```python
df2 = df1.rename(columns={'Group':'Country', 'G':'Gold','S':'Silver', 'B':'Bronze' })
df2
```

```python
df3 = df2.set_index(['Country']).stack().reset_index()
```

```python
df4 = df3.rename(columns={'level_1':'Medal Colour', 0:'Total', })
df4
```

```python
df4.to_csv('summerolympics.csv')
```
